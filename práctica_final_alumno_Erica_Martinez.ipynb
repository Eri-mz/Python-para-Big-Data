```python
from functools import reduce

'''
13. Dada una lista de Artículos con sus precio. Define las siguientes funciones:
Puedes definir más funciones si lo consideras necesario.
'''
articulos = {
    'Camisa': 20,
    'Pantalón': 30,
    'Calcetines': 5,
    'Zapatos': 50,
    'Gorra': 10,
    'Bufanda': 15,
    'Gafas': 25,
    'Reloj': 35,
    'Corbata': 40,
}

compra = ['Camisa', 'Pantalón', 'Pantalón', 'Gorra', 'Gafas', 'Corbata']
```

```python
# A. Una función que calcule el precio total de la compra
def suma(articulos,compra):
  precio = 0
  for index in compra:
    precio += articulos[index]
  return precio
suma(articulos,compra)
```

```python
# B. Una función que calcule el precio total de la compra con un descuento del 10%
def superior_100(articulos,compra):
  precio = 0
  for index in compra:
    precio += articulos[index]
  return precio *  0.90

superior_100(articulos,compra)
```

```python
# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€

def superior_100(articulos,compra):
  precio = 0
  for index in compra:
    precio += articulos[index]
  if precio > 100:
    precio*=0.9
  return precio

superior_100 (articulos,compra)
```

```python
# D. Una función que calcule el precio total aplicando el IVA (21%)
def precio_IVA(articulos,compra):
  precio = 0
  for index in compra:
    precio += articulos[index]
  precio *= 1.21
  return precio
precio_IVA(articulos,compra)
```

```python
#E.lista los artículos cuyo precio es superior a 20€
type(articulos.keys())
def precio_superior_20(articulos):
  superior20 = []
  for key, value in articulos.items():
    if value > 20:
      superior20.append(key)
  return superior20
precio_superior_20(articulos)
```
